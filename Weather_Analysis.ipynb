```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.pylab as pylab
```

```python
weather = pd.read_csv("1. Weather Data.csv")
weather.head()
```

```python
weather['Date'] = pd.to_datetime(weather['Date/Time']).dt.date
weather['Time'] = pd.to_datetime(weather['Date/Time']).dt.time
weather
```

```python
weather['Season'] = weather['Date'].apply(lambda x: 'Spring' if x.month in [3, 4, 5] else
                                     'Summer' if x.month in [6, 7, 8] else
                                     'Autumn' if x.month in [9, 10, 11] else
                                     'Winter')
weather
```

#### Rename the column name 'Weather' of the dataframe to 'Weather Condition'.

```python
weather.rename(columns={'Weather': 'Weather Condition'}, inplace=True)
weather
```

#### Q. 8 What is a correlation between 'Tem_C' , 'Rel Hum_%', 'Wind Speed_Km/h', & 'Visibilty_Km' and create a heat map?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation between temperature and relative humidity
correlation_temp_rh = weather['Temp_C'].corr(weather['Rel Hum_%'])

# Calculate correlation between wind speed and visibility
correlation_wind_speed_visibility = weather['Wind Speed_km/h'].corr(weather['Visibility_km'])

# Create a heatmap to visualize the correlations
correlation_matrix = weather[['Temp_C', 'Rel Hum_%', 'Wind Speed_km/h', 'Visibility_km']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

# Print the correlation values
print("Correlation between temperature and relative humidity:", correlation_temp_rh)
print("Correlation between wind speed and visibility:", correlation_wind_speed_visibility)
```

##### The values represent the strength and direction of the linear relationship between the variables. Here's what each correlation value means:
Correlation between temperature and relative humidity: -0.2201819230919853

The correlation coefficient of -0.2201819230919853 indicates a weak negative correlation between temperature and relative humidityThis means that as the temperature tends to increase, the relative humidity tends to decrease slightly, and vice versa. However, the relationship is not very strong.

Correlation between wind speed and visibility: 0.004882697258202263

The correlation coefficient of 0.004882697258202263 indicates a very weak positive correlation between wind speed and visibility.This means that there is almost no relationship or correlation between wind speed and visibility. The change in wind speed does not correspond to any significant change in visibility, and vice versa.

In summary, the correlation values near zero suggest that the variables (temperature, relative humidity, wind speed, and visibility) have little to no linear relationship with each other. The magnitude of the correlation coefficients close to zero indicates that any association between these variables is weak or negligible.

#### Q. 9What is a avegrage wind speed by each season?

```python
avg_wind_speed_by_season = weather.groupby('Season')['Wind Speed_km/h'].mean()
print("Average wind speed by season:")
print(avg_wind_speed_by_season)
```

The result displays the average wind speed (in km/h) for each season based on the 'Wind Speed_km/h' column in the dataset. Autumn and spring have relatively similar average wind speeds, with autumn averaging around 14.53 km/h and spring at approximately 14.88 km/h. These seasons experience moderate wind conditions, not too calm or too windy. Summer has a slightly lower average wind speed of around 13.49 km/h, indicating relatively calmer wind conditions. In contrast, winter stands out with the highest average wind speed among the seasons, averaging around 16.90 km/h. This suggests that winter experiences stronger and more gusty winds compared to the other seasons. The result provides valuable insights into the variations in wind speed across the different seasons, allowing us to better understand the typical wind patterns and conditions throughout the year.
